### Exercise 1: basic pct_change()

```python
import pandas as pd
s = pd.Series([50, 55, 52, 60, 58])
print(s.pct_change().mul(100).round(2))
```

### Exercise 2: rolling mean

```python
print(s.rolling(3, min_periods=1).mean().round(2))
```

### Exercise 3: groupby() + transform()

```python
df = pd.DataFrame({
    "ticker": ["A","A","A","B","B","B"],
    "close":  [10, 12, 11, 50, 55, 52]
})
df["rolling_2d"] = df.groupby("ticker")["close"].transform(
    lambda x: x.rolling(2, min_periods=1).mean()
)
print(df)
```
